This notebook demonstrates a simple linear regression analysis using [Python] to model Salary based on Years of Experience.

Read the dataset

```python
import pandas as pd
dataset = pd.read_csv("regression_data-1.csv")
dataset.head()
```

Create a scatter plot

```python
import matplotlib.pyplot as plt
plt.scatter(dataset["YearsExperience"], dataset["Salary"], color="red")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

Fit a linear model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(dataset[["YearsExperience"]], dataset[["Salary"]])
```

Overlay the regression line

```python
plt.scatter(dataset["YearsExperience"], dataset["Salary"], color="red")
plt.plot(dataset["YearsExperience"], model.predict(dataset[["YearsExperience"]]), color="blue")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

Evaluate the model: MSE, r, slope, intercept, and r_squared

```python
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error

x = dataset["YearsExperience"]
y = dataset["Salary"]
y_pred = model.predict(dataset[["YearsExperience"]]).ravel()

slope, intercept, r_value, p_value, std_err = linregress(x, y)
mse = mean_squared_error(y, y_pred)
r_squared = model.score(dataset[["YearsExperience"]], dataset[["Salary"]])

metrics = pd.DataFrame({
    "Metric": ["MSE", "r", "slope", "intercept", "r_squared"],
    "Value": [mse, r_value, slope, intercept, r_squared],
})
metrics
```

Plot with slope formula and correlation (r) in the top left corner

```python
plt.scatter(x, y, color="red")
plt.plot(x, y_pred, label="Fitted line", color="blue")
plt.text(
    0.05, 0.95,
    f"y = {slope:.2f}x + {intercept:.2f}\nr = {r_value:.2f}",
    transform=plt.gca().transAxes,
    fontsize=12,
    verticalalignment="top",
)
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
