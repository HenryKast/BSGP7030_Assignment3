This notebook demonstrates a simple linear regression analysis using [R] to model Salary based on Years of Experience.

Read the dataset

```r
dataset <- read.csv("regression_data-1.csv")
head(dataset)
```

Create a scatter plot

```r
plot(dataset$YearsExperience, dataset$Salary, col = "red",
     main = "Salary vs Experience",
     xlab = "Years of Experience",
     ylab = "Salary")
```

Fit a linear model

```r
model <- lm(Salary ~ YearsExperience, data = dataset)
model_summary <- summary(model)
model_summary
```

Overlay the regression line

```r
library(ggplot2)
ggplot() +
  geom_point(aes(x = dataset$YearsExperience, y = dataset$Salary), colour = "red") +
  geom_line(aes(x = dataset$YearsExperience, y = predict(model, newdata = dataset)), colour = "blue") +
  ggtitle("Salary vs Experience") +
  xlab("Years of experience") +
  ylab("Salary")
```

Evaluate the model: MSE, r, slope, intercept, and r_squared

```r
slope <- coef(model)[["YearsExperience"]]
intercept <- coef(model)[["(Intercept)"]]
r_squared <- model_summary$r.squared
r_value <- cor(dataset$YearsExperience, dataset$Salary)
y_pred <- predict(model, newdata = dataset)
mse <- mean((dataset$Salary - y_pred)^2)

metrics <- data.frame(
  Metric = c("MSE", "r", "slope", "intercept", "r_squared"),
  Value = c(mse, r_value, slope, intercept, r_squared)
)
metrics
```

Plot with slope formula and correlation (r) in the top left corner

```r
annotation_label <- paste0(
  "y = ", sprintf("%.2f", slope), "x + ", sprintf("%.2f", intercept),
  "\nr = ", sprintf("%.2f", r_value)
)

ggplot() +
  geom_point(aes(x = dataset$YearsExperience, y = dataset$Salary), colour = "red") +
  geom_line(aes(x = dataset$YearsExperience, y = y_pred), colour = "blue") +
  annotate(
    "text",
    x = -Inf,
    y = Inf,
    label = annotation_label,
    hjust = -0.05,
    vjust = 1.2,
    size = 4
  ) +
  ggtitle("Salary vs Experience") +
  xlab("Years of experience") +
  ylab("Salary")
```
